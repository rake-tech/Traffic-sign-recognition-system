1.2 Introduction To Resnet50

ResNet (Residual Network-50) is a deep Convolutional Neural Network (CNN) architecture introduced by researchers at Microsoft Research in 2015. It consists of 50 trainable layers and was designed to overcome the vanishing gradient problem, which makes training very deep neural networks difficult. ResNet-50 introduces the concept of residual learning through skip (shortcut) connections, allowing information and gradients to flow directly across layers. This enables the network to learn more effectively while maintaining high accuracy, even as the network depth increases.

ResNet-50 is capable of extracting rich and hierarchical features from images. In the initial layers, it learns low-level features such as edges, lines, and textures. As the data passes through deeper layers, the network captures more complex patterns, including shapes, objects, and semantic information. This hierarchical feature learning makes ResNet-50 highly effective for a wide range of computer vision applications.

A major advantage of ResNet-50 is its support for transfer learning. The model is commonly pre-trained on the ImageNet dataset, which contains over one million images across 1,000 categories. Instead of training a new model from scratch, developers can fine-tune the pre-trained ResNet-50 for specific tasks using smaller datasets. This approach reduces training time, lowers computational cost, and often improves model accuracy.

3.5 Label Encoding

The class labels were converted into one-hot encoded vectors using the to_categorical() function from Keras.
```python
import os
import cv2
import numpy as np

data = []
labels = []
classes = 43

# CORRECT PATH AFTER EXTRACTION
base_path = "/content/GTSRB"

for i in range(classes):

    path = os.path.join(base_path, 'train', str(i))  #Creates path to current class folder.

    images = os.listdir(path)

    for a in images:

        try:
            img_path = os.path.join(path, a)
            img = cv2.imread(img_path)

            img = cv2.resize(img, (30, 30))

            data.append(img)

            labels.append(i)

        except Exception as e:
            print("Error loading image:", a)
            print(e)

data = np.array(data)
labels = np.array(labels)

print("Data Shape:", data.shape)
print("Labels Shape:", labels.shape)
```

4. Transfer Learning using ResNet
4.1 Introduction to Transfer Learning

Transfer Learning is a deep learning technique in which a pre-trained model is reused for a new but related task. Instead of training a neural network from scratch, a model that has already learned features from a large dataset is adapted to solve a specific problem. This approach reduces training time, requires less data, and often achieves higher accuracy. In this project, Transfer Learning was used with the ResNet50 architecture to classify traffic sign images from the GTSRB dataset more efficiently and accurately.

### 4.2 Introduction to ResNet50

ResNet50 (Residual Network 50) is a deep convolutional neural network consisting of 50 layers that is widely used for image classification tasks. It was developed by Microsoft Research and won the ImageNet Large Scale Visual Recognition Challenge (ILSVRC). The main advantage of ResNet50 is its use of residual or skip connections, which help overcome the vanishing gradient problem and enable the training of very deep networks. ResNet50 is pre-trained on millions of images from the ImageNet dataset and can be used for transfer learning to solve new image classification problems with high accuracy. In this project, ResNet50 is used as a feature extractor to classify traffic sign images from the GTSRB dataset into 43 different categories.


### 4.3 ResNet50 Architecture

The ResNet50 architecture used in this project consists of a pre-trained ResNet50 model followed by custom classification layers. The ResNet50 model acts as a feature extractor by identifying important visual patterns such as edges, shapes, textures, and traffic sign characteristics from the input images. After feature extraction, a Global Average Pooling layer is used to reduce the dimensionality of the feature maps. A Dense layer with 256 neurons and ReLU activation is added to learn task-specific features, followed by a Dropout layer with a rate of 0.5 to reduce overfitting. Finally, a Dense output layer with 43 neurons and Softmax activation is used to classify the traffic sign images into one of the 43 categories present in the GTSRB dataset.

### 4.4 Freezing Pretrained Layers

In this project, the pre-trained layers of the ResNet50 model were frozen by setting their trainable parameter to False. This means that the weights learned from the ImageNet dataset were retained and not updated during training. Freezing the pretrained layers helps preserve the powerful feature extraction capabilities of ResNet50 while reducing training time and computational cost. Only the newly added classification layers were trained on the GTSRB traffic sign dataset, allowing the model to adapt to the traffic sign recognition task without losing the knowledge gained from large-scale image training.


### 4.5 Adding Custom Classification Layers

After freezing the pretrained ResNet50 layers, custom classification layers were added to adapt the model for the traffic sign classification task. A Global Average Pooling layer was used to convert the extracted feature maps into a compact feature vector. This was followed by a Dense layer with 256 neurons and ReLU activation to learn traffic sign-specific features. A Dropout layer with a rate of 0.5 was included to reduce overfitting by randomly disabling neurons during training. Finally, a Dense output layer with 43 neurons and Softmax activation was added to classify images into the 43 traffic sign categories of the GTSRB dataset.


```python
#Building the model
model = Sequential()
#Rectified Linear Unit (ReLU) f(x)=max(0,x)

# Input Layer
model.add(Input(shape=X_train.shape[1:]))        #image dimensions for CNN

# First Convolutional Layer
model.add(Conv2D(filters=32, kernel_size=(5, 5), activation='relu')) #extracts basic image features

# Second Convolutional Layer
model.add(Conv2D(filters=32, kernel_size=(5, 5), activation='relu'))

# Max Pooling Layer
model.add(MaxPooling2D(pool_size=(2, 2)))     #Reduce image size.

# Dropout Layer
model.add(Dropout(rate=0.25))          #randomly disabling neurons.
# Third Convolutional Layer
model.add(Conv2D(filters=64, kernel_size=(5, 5), activation='relu'))

# Fourth Convolutional Layer
model.add(Conv2D(filters=64, kernel_size=(5, 5), activation='relu'))  #captures more complex image structures.

# Max Pooling Layer
model.add(MaxPooling2D(pool_size=(2, 2)))

# Dropout Layer
model.add(Dropout(rate=0.25))

# Flatten Layer
model.add(Flatten()) #converts multidimensional data into a single vector

# Fully Connected Layer
model.add(Dense(256, activation='relu'))

# Dropout Layer
model.add(Dropout(rate=0.5))
# Output Layer
model.add(Dense(43, activation='softmax'))
model.summary()
```

```python
# Building the model (re-added for robustness against kernel state issues)
model = Sequential()
model.add(Input(shape=X_train.shape[1:]))
model.add(Conv2D(filters=32, kernel_size=(5, 5), activation='relu'))
model.add(Conv2D(filters=32, kernel_size=(5, 5), activation='relu'))
model.add(MaxPooling2D(pool_size=(2, 2)))
model.add(Dropout(rate=0.25))
model.add(Conv2D(filters=64, kernel_size=(5, 5), activation='relu'))
model.add(Conv2D(filters=64, kernel_size=(5, 5), activation='relu'))
model.add(MaxPooling2D(pool_size=(2, 2)))
model.add(Dropout(rate=0.25))
model.add(Flatten())
model.add(Dense(256, activation='relu'))
model.add(Dropout(rate=0.5))
model.add(Dense(43, activation='softmax'))

# Compile the model (re-added for robustness against kernel state issues)
model.compile(loss = 'categorical_crossentropy',optimizer = 'adam',metrics = ['accuracy'])

epochs = 20
history = model.fit(X_train,y_train,batch_size =32,epochs = epochs,validation_data= (X_test,y_test))
```
